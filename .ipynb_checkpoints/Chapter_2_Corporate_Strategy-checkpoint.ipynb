## 1. Linking Valuation to Strategy: The Origin of the Value


### 1.1. Differentiation with Unique, Innovative Products

#### 1.1.1. Stragtegy

**Aim:**
Achieving proprietary knowledge, coupled with a marketing program for customer acceptance.

**Important in:**
Industries involving innovative or uniqueproducts, e.g., pharmaceutical industry, information technology and electronics.


#### 1.1.2. Stragtegic Growth Option Value

**Portfolio of Real Options:**
- The strategy encompasses a portfolio mix of compound options, 
    - e.g., R&D followed by prototyping pilot projects in new markets. 

- High technical uncertainty over success of R&D, due to implicit leverage, coupled with high commercial uncertainty and strategic uncertainty over product acceptance. 

- Adaptive capabilities enhance the value of flexibility.


**Strategic Moves (Game Theory):**

- Commitment effect of first mover advantages due to e.g., patents, buyer switching costs and network externalities. 

- When the product has a network externality, an early mover that made more sales than its competitors in early periods develops a larger installed base. 

- In many cases the costs of imitation have become lower than the cost of innovation. However, increased technological intensity coupled with shorter life cycles makes preventing early imitation increasingly difficult. 

### 1.2. Leverage of Reputation

#### 1.2.1. Stragtegy

**Aim:**
Creating a leading position in quality or service, differentiating the product from competitors

**Important in:**
Industries with large brand-name capital, e.g., softdrinks or cigarette industry. The platform company should be a respected company to be a successful foothold for future growth opportunities.


#### 1.2.2. Stragtegic Growth Option Value

**Portfolio of Real Options:**
- Strategic/ goodwill investments in advertising and marketing generate future expansion options. 

- Low technological and commercial uncertainty.


**Strategic Moves (Game Theory):**
- Commitment effect may depend on position of the firm and competitive reaction. 

- Reputation and buyer choice under uncertainty may also be isolating mechanisms for the early-mover. 

    - If buyers have imperfect information regarding product quality, they may hold on to the first brand that performs satisfactory. 

- Threat of reciprocating reactions: advertising may result in more advertising and price competition may make everybody worse off.


### 1.3. Cost Advantage based on Economies of Scope

#### 1.3.1. Stragtegy

**Aim:**
Expansion is aimed at gaining cost advantages associated with producing and selling multiple products related by a common technology, product facilities or network.

**Important in:**
Industries where a cost advantage exists when the same investment can support multiple profitable activities at different locations.


#### 1.3.2. Stragtegic Growth Option Value

**Portfolio of Real Options:**
- Options that allow switching between different technologies or products and leverage of competences onto a broad geographical or financial base. 
    - For instance, R&D generating compound options resulting from critical technologies that cut across businesses. Various simple, commercial options over a broad product line.


**Strategic Moves (Game Theory):**
- An advantaged strategic position results from resources that can be used in several products. As a result, a given resource position will often have consequences for several products, each yielding part of the resulting return.


### 1.4. Build up Scale

#### 1.4.1. Stragtegy

**Aim:**
Building size in a fragmented market is aimed at market leadership with an investment and pricing policy that fully exploits economies of scale in the firmâ€™s functions. 

**Important in:**
Fragmented industries with large fixed investments in R&D or produc-
tion, banking, automobiles, oil.


#### 1.4.2. Stragtegic Growth Option Value

**Portfolio of Real Options:**
- Infrastructure investments that generate options to expand more quickly than competitors. This resource position enables the firm to preempt expansion opportunities in the market.



**Strategic Moves (Game Theory):**
- Value of early commitment is present when the minimum efficient scale is large relative to the market size and a limited number of firms can fit in the market without creating overcapacity. 
- Capacity games often involve contrarian reactions. 
- Late movers would be reluctant to pay for the resource since they would be faced with higher cost or might face the threat of price competition due to excess capacity.


### 1.5. Absolute Cost Advantage

#### 1.5.1. Stragtegy

**Aim:**
Expansion is aimed at achieving the lowest delivered cost position in the industry, especially if cost reduction can be made proprietary to the firm.


**Important in:**
Industries where proprietary knowledge, a learning cost effect, efficient supply or favorable locations result in a cost advantage, e.g., natural resource industries

#### 1.5.2. Stragtegic Growth Option Value

**Portfolio of Real Options:**
- Early exercise of options may generate a cost advantage or experience curve effects. 
    - For instance, acquisition of favorable locations or exploration investments in the petroleum industry to acquire favorable areas at low cost.



**Strategic Moves (Game Theory):**
- Early commitment of capital (exercise of real option) to acquire a cost advantage or experience effect. 

- Capitalizing on expansion opportunities in case of experience effects depends on the ability of the firm to keep experience effects proprietary. 

- When experience effects are proprietary, later resource producers will face an uphill battle with earlier producers who have lower costs. 

- Later acquirers would thus face less valuable expansion opportunities

## 2. Estimate Growth Option Value


$$
\text{Market Value} = \text{Growth Option Value} + \text{Value of Assets in place}
$$

Or precisely:

$$
\text{Equity Value} = \text{Value of Assets in place} + \text{Present Value of Growth Options (PVGO)}
$$
- Equity value: Known from financial market
    - Equity Value = No. of floating shares $\times$ share prices
- Value of assets in place: PV of earnings generated by the assets in place.
    - Estimated value of asset in place: Constant earnings profitability of current assets under no growth policy

- PVGO: "Back-solve" for the value of the growth opportunities from the stock price



### Example 2.1.

In 2012, just after the acquisition of Instagram, Facebook went public through an Initial Public Offering (IPO). Please explain two approaches for determining the growth option value of Facebook. Which approach do you prefer and why?


**Answer**

The answer discusses 2 methods for determining the growth option value and explains why it prefers one of the two methods.

For example:

- Top-down approach, we take the market value and subtract from that the value of assets in place, which leaves us with the growth option value 

- Bottom-up approach: Identify each option individually and calculate the value of the portfolio of options.

The top-down approach can only be used for traded companies (thus after an IPO). A **disadvantage of using the top-down approach** is that the market can be overvalued and therefore **overestimate the growth option value**. However, in the bottom-up approach the identification of all options can be difficult and lengthy process.

### Example 2.2

Please provide two reasons why growth firms (e.g., leading firms in information technology, pharmaceuticals, and consumer electronics) tend to have a higher option value component (PVGO) than income stocks.

**Answer**

The answer mentions two characteristics of growth firms: 1) compound options and 2) volatility. Further it explains why these two characteristics lead to a higher option value.

For example:

First, they tend to operate in more volatile and rapidly evolving industries (characterized by more frequent technological innovations and a more intensely competitive environment), with the higher underlying volatility being translated into higher (simple) option value.

Second, they tend to have a higher mix of compound (multi-stage or growth) options than simple (cash-generating) options, which further levers up (amplifies) their option value (being options on options).

### Example 3.1.3.

Please explain the difference between structural entry barriers and strategic entry barriers and give an example of both.

**Answer**

The answer explains that structural entry barriers are related to the 'nature' of the industry or company, while strategic entry barriers relate to specific actions taken by the company.

For example:

Structural entry barriers occur when established firms possess cost, marketing, location, or distributional advantages over potential entrants.

When incumbents take deliberate actions to deter the entrance of new competitors, barriers are said to be strategic.

Such actions include manipulating prices before entrance (limit pricing) or after entrance (predatory pricing), expanding capacity, obtaining patents (even for small variants of a product), advertising aggressively to create brand loyalty, and so on.

### 3.2. External Framework: Game Theory


Game theory looks at strategic interactions between firms. So the profitability of the firm depends on how it plays the game in its external environment against rivals. 





While game theory understands that the external environment, and how it plays the game in an external environment, it doesn't analyze so much the differences between firms. For instance, the strategic conflict and game theory approach does not recognize that the accumulation of unique resources and unique capabilities is also an important determinant of profitability. 

### 3.3. Internal Framework: Resource-based View

The resource based view states that the accumulation of unique resources, and capabilities are the source of competitive advantage. 

Resources are actually the basic input in the the production process, and they add to the strengths and weaknesses of the firm, e.g. trade contracts, the technical knowledge or know-how.

A capability is a capacity to perform certain tasks in the production process, we also speak of core capabilities, e.g.: 
- Tesla: know-how in the electric cars
- Canon: know how in optics. 

To be a source of competitive advantage, these resources and capabilities should have several characteristics. 
1. should be **relevant** 
2. should be **hard and durable**, **difficult to imitate and replicate**
3. The value created by the resource should be **appropriable** by the firm. 

If these conditions are met, the firm can expect to earn more than the opportunity cost of capital that adds to the value assets in place or can expect valuable growth opportunities.

## 4. Economic Valud Added (EVA)

### 4.1. Definition and Interpretation

Economic Valud Added (EVA)
- Definition: The value creation from the funds invested into the business. 
- Direct Calculation: $(\text{ROCB} - \text{WACC}) \times \text{Operating Capital}$
- Implication: 
    - Growth, by itself, does NOT create value.
    - Growth, with investments in excess return, generates value. 
- Interpretation:
    - a positive EVA: a firm creates value from the invested funds
    - a negative EVA: the firm is not generating any value.


### 4.2. Calculation

$$
\begin{align*}
\text{EVA} &= (\text{ROCB} - \text{WACC}) \times \text{Operating Capital} \\
&= \left(\frac{\text{EBIT} \times (1 - \tau_c)}{\text{Operating Capital}} - \text{WACC}\right) \times \text{Operating Capital} \\
&= \text{EBIT} \times (1 - \tau_c) - \text{WACC} \times \text{Operating Capital} \\
&= \text{NOPLAT} - \text{Capital Charge}
\end{align*}
$$

where:
- $\text{ROCB}$: Return on the firmâ€™s operating capital
- $\text{NOPLAT}$: Operating profits after taxes
- $\text{Operating Capital}$: 
    - the firmâ€™s invested capital (i.e. the sum of all operating working capital (current operating assets minus current operating liabilities), 
    - fixed assets, 
    - intangible assets, and
    - net other long-term operating assets (net of long-term operating liabilities)).
- $\text{Capital Charge}$: Return investors demand on the firm's operating capital
- $\tau_c$: tax rate


### 4.3. Analysis over time



1. Early development: 
    - Value destruction, i.e. ROCB is generally lower than its WACC.
    - involve firms that have just penetrated the market and thus generate low returns 
    - while their investors demand high compensation for the companyâ€™s risk. 


2. After a certain period: 
    - gain market power and establish a stable investment pattern
    - ROCB increases and WACC decreases
    

3. In this late development stage
    - The firm creates value as its operations grow.


4. However, competitors are attracted as the economic profits rise
    - the firmâ€™s ROCB fades and its value creation decreases. 


5. Finally, 
    - Able to maintain a strong competitive advantage
        - ROCB declines until it equals the WACC, i.e. no value creation 
    - Able to maintain a strong competitive advantage: 
        - the incremental difference between the ROCB and WACC can stay positive.

### 4.4. Advantages and Disadvantages

#### 4.1. Advantage

- A clear indication of whether a firm creates value from the capital invested in the business

- Managers are forced to take both the firmâ€™s assets and expenses into account
    - since the firmâ€™s balance sheet items are included in the measureâ€™s computation
    
    
#### 4.2. Disadvantage

- Difficult to apply to companies with intangible assets
    - e.g. companies with patents or a rich technology base
    - Intangible assets are hard to value
    
- Operating Capital subjective to different accounting treatments
