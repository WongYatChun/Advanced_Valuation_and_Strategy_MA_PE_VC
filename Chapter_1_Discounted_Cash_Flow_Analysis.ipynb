### Variations of DCF methods

1. Weigted Average Cost of Capital (WACC) method
    - Discount cash flows of the firms with wieghted average of cost of capital
    
    
2. Free Cash Flow to Equity (FCFE) method
    - Discount cash flows to equity with the cost of equity (to be used when valuing listed companies and financial institutions
    
    
3. Adjusted Present Value (APV) mehthod
    - Discount cash flows of an all-equity firm and add the value of the financing side effects
    - appropriate when the debt-equity mix is changing over time

### DCF Framework

The company's expected cash flows can be separated into two periods:

1. The expected forecast period
    - typically span up to six years into the future


2. The terminal or continuing value
    - cover the period after the explicit forecast period

**Enterprise value = PV of FCF during the forecast period + PV of the terminal value**

### Terminal Value

Since the terminal value amounts for 50% to 125% of the total value, a thoughtful estiamte of terminal value is very important

#### 1. Perpetuity Method: 

Links cash flows to growth rates, reinvestment rates, and return on capital base, e.g.
- Key value driver Method
- Convergence Model
- Gordon Growth Model


$\text{NOPAT}$: After-tax operating profit at time $T$

$\text{ROCB}$: Return on invested capital for new investment

$g$: Expected long-term growth rate in revenue & cash flows

$\text{WACC}$: The weighted average cost of capital, based on long-run target capital structure

#### Aggresive (Gordon) Growth Model:

$$
\frac{\text{NOPAT}_t \times (1 + g)}{\text{WACC} - g}
$$

It assumes that the returns on new investments ($\text{ROCB}$) approach infinity. It is usually unrealistic and seldomly used.


#### Key Value Driver Method:

$$
\frac{\text{NOPAT}_t \times (1 + g) \times \left(1 - \frac{g}{\text{ROCB}}\right)}{\text{WACC} - g}
$$

It assumes that the returns on new investments ($\text{ROCB}$) are higher than the WACC. It should be used if the company is expected to maintain their competitive advantage consistently, e.g. Coca Cola.


#### Convergence Model:

$$
\frac{\text{NOPAT}_t \times (1 + g)}{\text{WACC}}
$$

It assumes that the returns on new investments ($\text{ROCB}$) are equal to the WACC. It should be used in competitive industries that move to a competitive equilibrium, e.g. Most firms.

### Example



```python
NOPAT = 117
g = 0.03
ROCB = 0.1
WACC = 0.09

KVD = NOPAT * (1 + g) * (1 - g/ROCB) / (WACC - g)
round(KVD,0)
```

```python
pv_KVD = KVD * (1 + WACC)**-(2024 - 2018)
round(pv_KVD,0)
```

Assume that a company will be worth some multiple of future earnings or book value in the continuation period, e.g.

- P/E Ratio
- EV/EBITDA

Typically, practitioners apply multiples to EBIT or EBITDA. EBITDA metric ignores capital structure, and therefore can be used between similar companies in the industry, regardless of their debt-to-EBITDA ratios.

### Terminal Value Check

In cases where the free cash flows are low, the valuation almost comes down entirely to continuing value estimations. Executives must therefore perform checks and corrections of the estimations of each of the components making up the terminal value, to guard against overvaluation. Several important checks include the following: 

1. The choice of the value-driver formula has to be based on the industry and company analysis of the long-term profitability.
    - Check whether the formula used for calculating the terminal value is congruent with the situations/scenarios of the company
    
    
2. The extrapolation of historical cash flows during the growth or forecasting period
    - Steps to estimate of the growth rate $g$ after the planning period:
        1. $\text{Net investments} = \text{Investment in the NWC} + \text{total capex} - \text{decreciation}$
        2. $\text{Reinvestment Ratio} = \text{Net Investment} \div \text{NOPAT}$
        3. $g = \text{ROCB} \times \text{Reinvestment Ratio}$
    - Use averages of $\text{NOPAT}$ and growth for cyclical industries
        

3. The consistency on the entry and exit (or terminal) multiple
    - Common in private equity valuations
    - relate the exit multiple of the DCF calculation at the terminal date at which a company can be sold, to the entry multiple for which it would be acquired.
    - Many financial parties consider multiple arbitrage
        - Value Creation: the entry multiple is smaller than the exit multiple assuming that the company can be sold for more than it is acquired. 
        - Multiple arbitrage is unlikely if deal markets are already hot and pricing is at its peak, unless there are clear options for improving the inherent value of the acquired business, e.g. with restructuring or the creation of new growth opportunities, or when there are new sources of value for the buyer at exit.

### Alternative to WACC: APV Method

The WACC method assumes that the company will be financed with a constant debt/equity ratio. Although the effects of a time varying D/E ratio can be incorporated in a recursive WACC, it is much easier and more accurate to use the APV method.


The unlevered film value of the company:
- calculated in the same way as the enterprise value is calculated in the WACC method, only now the unlevered cost of capital is used as discount factor instead of the WACC.



According to the Adjusted Present Value-method, tax shields on interest have to be valued separately. One could either use the unlevered cost of capital or alternatively the cost of debt as a discount rate for valuing the tax shields on interest. However, if the debt levels are determined using a multiple of EBITDA. Debt ratios are assumed to be predetermined and time varying and the debt level is related to firm value. The debt level therefore varies with the market value and the risk of the tax shields are related to the risk of the business. Therefore, in this case, unlevered cost of capital is more appropriate as the discount factor.


The present value from the debt induced tax shields

1. Calculate the interest costs for every year of the planning period. 
    - The interest costs for a certain year are calculated over the average debt level of the current and previous year. 
    - Use the debt multiple to obtain the debt levels and use the sum of the risk free rate and credit spread debt as interest percentage.
    
    
2. Calculate the tax shields for the planning period and discount these with the appropriate discount factor.


3. Calculate the terminal value (TV) of tax shields in a similar way as the terminal value is calculated in the WACC method, using the appropriate discount factor instead of the WACC. Use the following formula 
    - $\text{TV} = \frac{\text{Tax Shield} \times (1+g)}{\text{Discount factor} - g}$


4. Finally add together the present value of tax shields and the present value of the terminal value of tax shields to obtain the present value of debt induced tax shields.

## Internal Rate of Return (IRR)

The IRR is the discount rate which makes the Net Present Value (NPV) of all cash flows from an investment or project equal to zero.

$$
0 = \sum^{T}_{t=0} \frac{\text{CF}_t}{1 + \text{IRR}}
$$

The IRR is also used as proxy for the growth rate that a project is expected to display. In practice an investment is considered profitable when its IRR is higher than its cost of capital or the Required Rate of Return. 


### IRR and Private Equity

The IRR can also be used to determine what maximally can be invested, for instance by a private equity firm.

Consider a private equity firm that wants to acquire a target company at $t = 0$. It will use \\$450 million of debt to fund the debt proportion related to the acquisition at the entry (acquisition) date. The private equity investor plans to exit after 5 years (at $t = 5$) and expects to realize an exit EBITA-multiple of 10. Because all intermediate cash flows the company will generate are used to repay the debt, the expected net debt at $t = 5$ will be \\$250 million. The EBITA of the company at $t = 5$ equals to \\$103 million. What is the maximum enterprise value and equity investment the private equity firm is willing to invest at the entry date when the private equity firm requires an IRR of 25%?



Note that if the private equity firm is able to finance the project at entry with \\$600 million debt, all else equal, the equity investment at entry would then be $705.59  600 = 105.59$, and the IRR would be
$(780/105.59)^{(1/5)} = 49\%$.

```python
import numpy as np
import numpy_financial as npf

exit_equity_value = 976
entry_equity_value = -362
year = 2024-2018
cf = np.array([0]*(year+1))
cf[0] = entry_equity_value
cf[-1] = exit_equity_value
cf
```

```python
irr = npf.irr(cf)
round(irr * 100, 1)
```
